```python
# ==========================================
# 1: Environment Setup & Model Loading
# ==========================================

from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

# Switching to TinyLlama for factual, assistant-like responses
model_name = "TinyLlama/TinyLlama-1.1B-Chat-v1.0"

print(f"Loading tokenizer and model for '{model_name}'... Please wait.")

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

print("Model and Tokenizer loaded successfully! Ready to chat.")
```

```python
# ==========================================
# 2: Chatbot Logic & Conversation Loop
# ==========================================

print("Chatbot starts and waits for input")
print("Chatbot: Hello! I am your AI assistant. How can I help you today?")
print("(Note: Type 'exit' or 'quit' to end the conversation)\n")

# 1. Initialize conversation history with the proper TinyLlama prompt format
# The </s> tells the model "this section is done"
chat_history = "<|system|>\nYou are a friendly, concise, and helpful AI assistant. Provide short, direct answers.</s>\n"

while True:
    user_input = input("User: ")

    if user_input.strip().lower() in ['exit', 'quit']:
        print("Chatbot ends the conversation")
        break

    # 2. Append the new user input to the history, including the </s> stop token
    chat_history += f"<|user|>\n{user_input}</s>\n<|assistant|>\n"

    inputs = tokenizer(chat_history, return_tensors="pt")

    # 3. Generate the response
    outputs = model.generate(
        **inputs,
        max_new_tokens=200,
        do_sample=True,
        temperature=0.3,
        top_p=0.9,
        pad_token_id=tokenizer.eos_token_id,
        eos_token_id=tokenizer.eos_token_id # Explicitly tell it to stop at the end token
    )

    # 4. Extract ONLY the newly generated tokens
    input_length = inputs["input_ids"].shape[1]
    new_tokens = outputs[0][input_length:]

    # 5. Decode the assistant's reply
    response = tokenizer.decode(new_tokens, skip_special_tokens=True).strip()

    print(f"Chatbot: {response}")

    # 6. Append the assistant's response to the history so it remembers for the next turn
    chat_history += f"{response}</s>\n"
```
